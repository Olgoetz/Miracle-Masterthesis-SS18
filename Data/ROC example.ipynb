```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import library to modify style parameters
from matplotlib import rcParams

# font styles
fontsizes= {'title' : 20, 'axis': 15, 'ticks': 13}
fontweights= {'title' : 800, 'axis' : 600}

# define styles for the images
rcParams['axes.titlepad'] = 20

# define the output resolution 
output_dpi = 300
```

```python
fig, ax = plt.subplots()

baseline_x = np.arange(0,1.1,0.1)
baseline_y = np.arange(0,1.1,0.1)


ax.set_xlim(left=0, right=1.0)
ax.set_ylim(bottom=0, top=1.0)
ax.plot(baseline_x, baseline_y, linestyle='--')


A = [0,0]
ax.plot(A[0], A[1], marker='x', color='black', markersize=13)
ax.annotate('A', xy=A, xytext=(A[0]+0.05, A[1]+0.15), arrowprops=dict(facecolor='black', 
                                                                      arrowstyle="-|>"), fontsize=13)

B = [0,1]
ax.plot(B[0], B[1])
ax.annotate('B', xy=B, xytext=(B[0]+0.05, B[1]-0.15), arrowprops=dict(facecolor='black', 
                                                                      arrowstyle="-|>"), fontsize=13)

C = [0.5,0.5]
ax.plot(C[0], C[1], marker='x', color='black', markersize=13)
ax.annotate('C', xy=C, xytext=(C[0]+0.01, C[1]+0.05), fontsize=13)

D = [1,1]
ax.plot(D[0], D[1])
ax.annotate('D', xy=D, xytext=(D[0]-0.15, D[1]-0.1), arrowprops=dict(facecolor='black', 
                                                                      arrowstyle="-|>"), fontsize=13)

E = [0.2,0.7]
ax.plot(E[0], E[1], marker='x', color='black', markersize=13)
ax.annotate('E', xy=E, xytext=(E[0]+0.01, E[1]+0.05), fontsize=13)

F = [0.6,0.92]
ax.plot(F[0], F[1], marker='x', color='black', markersize=13)
ax.annotate('F', xy=F, xytext=(F[0]-0.01, F[1]+0.03), fontsize=13)


graph1_x = np.arange(0,1.1,0.1)
graph1_y = [0,0.5,0.7,0.78,0.83,0.85,0.87,0.9,0.93,0.96,1]
ax.plot(graph1_x, graph1_y)

graph2_x = np.arange(0,1.1,0.1)
graph2_y = [0,0.5,0.65,0.7,0.80,0.81,0.92,0.94,0.97,0.995,1]
ax.plot(graph2_x, graph2_y)

#ax.set_title('ROC Illustration', fontdict=dict(fontsize=fontsizes['title'], fontweight=fontweights['title']))
ax.set_xlabel('False Positive Rate', fontdict={'fontsize' : fontsizes['axis'], 'fontweight' : fontweights['axis']})
ax.set_ylabel('True Positive Rate', fontdict={'fontsize' : fontsizes['axis'], 'fontweight' : fontweights['axis']})
ax.tick_params(axis='both', labelsize=fontsizes['ticks'])

# save the figure
plt.savefig('roc_example.png',dpi=output_dpi, bbox_inches='tight')
```
